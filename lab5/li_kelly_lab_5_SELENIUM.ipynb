```python
import selenium

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
```

```python
PATH = Service("/Users/vanilla/Downloads/chromedriver_mac64/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

website = "https://techwithtim.net"

driver.get(website)
# print(driver.title)
# driver.close()

search = driver.find_element(By.NAME, "s")

search.clear()
search.send_keys('test')
search.send_keys(Keys.RETURN)

try:
    main = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.ID,'main')))
    print(main.text)
    
    articles = main.find_elements(By.TAG_NAME, 'article')
    
    for i in articles:
        header = i.find_element(By.CLASS_NAME, "entry-summary")
        print(header.text)

except:
    driver.quit()
```

```python
import pandas as pd
PATH = Service("/Users/vanilla/Downloads/chromedriver_mac64/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

website = 'https://www.wikipedia.org'

driver.get(website)

search = driver.find_element(By.NAME, "search")

search.clear()
search.send_keys('cat')
search.send_keys(Keys.RETURN)

csv_file = open('wikicat.csv','w',newline='',encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Links to images"])

try:
    main = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CLASS_NAME,'mw-page-container-inner')))

    articles = main.find_elements(By.CLASS_NAME, "image")
    
    for i in articles:
        img_group = i.find_element(By.TAG_NAME, 'img')
        link = img_group.get_attribute('src')
        print(link)
        
        csv_writer.writerow([link])
    
    csv_file.close()

except:
    driver.quit()
```
