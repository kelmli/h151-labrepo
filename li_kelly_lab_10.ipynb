```python
# Quote GUI 1
import requests
from tkinter import *
import random

root = Tk()
root.title("11:11")
root.geometry("600x400")

url = "https://api.quotable.io/quotes/random?limit=20"
json_file = requests.get(url).json()

names = []
for i in range(0,len(json_file)):
    author = json_file[i]['author']
    names.append(author)
        
def showquote():
    
    name = value.get()
    result_label = Label(root, text=f"You've chosen {name}")
    result_label.pack()
    
    number = names.index(name)
    
    textbox.delete(0.0, END) 
    text=json_file[number]['content']
    textbox.insert(END,text)
    
value = StringVar()
value.set("Select a Name")
dropdown = OptionMenu(root, value, *names)
dropdown.pack()

label1 = Label(root, text= "Here is the quote", font = 18,padx=20, pady=20)
label1.pack()
textbox = Text(root, font="Arial", height=5,width=40,bg='pink',padx=30,pady=30)
textbox.pack()
show = Button(root, text="Show Quote", command=showquote)
show.pack(side=BOTTOM)

root.mainloop()

# When you close the widget and replay it, it will give you a whole new list.
```

```python
# Weather GUI 2
import requests
from tkinter import *

root = Tk()
root.title("Weather App")
root.geometry("700x500")

lat = "42.098701"
lon = "-75.912537"
genius = requests.get(f"https://api.weather.gov/points/{lat},{lon}")
json_file = genius.json()

json_file['@context']

forecast = json_file['properties']['forecast']
new_request = requests.get(forecast)

json = new_request.json()

#list box - new widget
Lb1 = Listbox(root, selectmode="single")

periods = json['properties']['periods']
n=0
temps = []
names = []
dtfs = []
for i in periods:
    name = i['name']
    names.append(name)
    Lb1.insert(n,name)
    temp = i['temperature']
    temps.append(temp)
    dtf = i['detailedForecast']
    dtfs.append(dtf)
    n+1
print(temps)

def confirm():
    day= Lb1.curselection()
    for i in day:
        chosen_day = Lb1.get(i)
    result_label = Label(root, text=f"You've chosen {chosen_day}")
    result_label.pack()
    
    number = names.index(chosen_day)
    t = temps[number]
    d = dtfs[number]
    
    label1 = Label(root, text= f"{t}Â°F", font=('Helvetica', 70),padx=20, pady=20)
    label1.pack()
    label2 = Label(root, text= d, font=('Helvetica', 15),padx=20, pady=20)
    label2.pack()

confirm = Button(root, text="Show Weather", command=confirm)
confirm.pack(side=BOTTOM)

Lb1.pack()

root.mainloop()
```
