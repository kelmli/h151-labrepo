```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance
    
    def wd(self,amt):
        self.balance -= amt
    
    def dp(self,amt):
        self.balance += amt
        
    def show(self):
        return self.balance
    
b1 = BankAccount(1000)
b1.wd(10)
b1.dp(1000)
b1.show()
    
```

```python
#3
#For this example, I used how many types of beans I have.

beans = {'lime':8, 'cherry': 10, 'blueberry':23, 'bubblegum':29, 'lemon':19}
#creating my dictionary. 

def eat(flavor, x):
    eaten = beans[flavor] - x
    beans.update({flavor: eaten})
    
    return beans
# subtracts x from the beans of that flavor because they have been eaten. 
# Then, it updates the value in the dictionary.

def buy(flavor, x):  
    if flavor in beans.keys():
        bought = beans[flavor] + x
        beans.update({flavor: bought})
        
    else:
        beans.setdefault(flavor, x)
    
    return beans
# adds x to the amount of beans from that flavor. Then it updates it.
# However, if the beans are not already in the dictionary, it inserts the new key with its specified value using setdefault().

def emptybeans():  
    return beans.clear() 
#if i were to clear my dictionary (or if I emptied out all the beans)
        
def count_beans(flavor):
    if flavor in beans.keys():
        print(f"There are {beans.get(flavor)} {flavor} beans.")
    else:
        print(f"We don't have any {flavor} beans.")
# counts how many beans I have if I have that flavor in the dictionary.
#.get() returns the value of the specified key.
# However, if the flavor is not in beans.keys(), it tells us that we have none of that flavor.

print(eat('lime', 3))
print(eat('cherry',1))

print(buy('choco', 10))
print(buy('lime', 100))
print(buy('lemon', 30))

count_beans('lemon')

print(emptybeans())
```

```python
#5
import random

from IPython.display import clear_output 
sowpods = "sowpods.txt"

def word_generator():
  filename = sowpods
  f = open(filename, "r", encoding="utf8")
  words = f.readlines()
  f.close()

  return random.choice(words).strip()

answer = word_generator()

#chossing a random word from the sowpods text file.
#source is from Dr.Gibson's version of Hangman.
'''https://colab.research.google.com/drive/165XxeQYWzz7SKucdGRyEs2Vl5XIEqW5Y?usp=sharing'''
#This importing method was refreshed during class, as I have not done this since HARP150's final data project.

def hangman():
    word = answer.lower()
    wordlist = []
    tries = 7

    for i in word:
        if i not in wordlist:
            wordlist.append(i)

    right = []
    wrong = []
    guesses = ' '
    
    #while loop so that tries don't get exceeded.
    #also so that it ends when the player gets the word. (letters in right list are the same in wordlist)
    while tries > 0 and right != wordlist :

        guess = input('guess a letter: ')
        guesses += guess # add the players guess to guesses via concatenation
        print("Letter Bank:", guesses) 

        if guess in right or guess in wrong:
            print('Already guessed this word, try again!')
            # if the player already guessed the word.
        
        else:
        #parts of this "for char in word" chunk below was sourced from this website:
        #https://www.pythonforbeginners.com/code-snippets-source-code/game-hangman
            for char in word:
                if char in guesses:   
                    print (char,end=""), #prints out the character if the guess is in the word.   
                          #end="" tells the print statement not to add a new line    
                else:
                    print ("_ ",end=""), 

            if guess in wordlist:
                print('\nGood!')
                right.append(guess) #adds it to the right list.

            else:
                wrong.append(guess)
                print('\nBad.')
                print(f'You have {tries-1} tries left')

                tries-=1 #subtracts a try every time player guesses wrong.
                if tries == 0:
                    print(f' The word was {word.upper()}.')
                    #reveals the word if there are no more tries.

        
        if len(right) == len(wordlist):
            print('You got the word! Yay!')
            break
        
hangman()

# I was refreshed on how to use functions and also the for,while, and if loops. 
# Also conditionals such as ==, >, and < were engaged.
```
