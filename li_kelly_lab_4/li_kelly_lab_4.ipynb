```python
from bs4 import BeautifulSoup
import requests
import html5lib
import csv
import pandas as pd
```

```python
source = requests.get("https://www.imdb.com/list/ls055592025/").text
imdb_soup = BeautifulSoup(source, "lxml")

csv_file = open('imdb_scrape.csv','w',newline='',encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Title','Year',"Rating"])

for element in imdb_soup.find_all('div', class_="lister-item-content"):
    title = element.h3.a.text
    year = element.find('span', class_ = "lister-item-year text-muted unbold").text.strip('()')
    stars = element.find('span',class_="ipl-rating-star__rating").text.strip()
    
    csv_writer.writerow([title,year,stars])
csv_file.close()
                                                                                    
```

```python
csv_file = open('cat_scrape.csv','w',newline='',encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Title','Item Description',"Webpage Hyperlink"])

page = 1

while page !=6:
    cat_source = requests.get(f'https://www.loc.gov/search/?q=cats&sp={page}').text
    cat_soup = BeautifulSoup(cat_source, "lxml")
    
    for element in cat_soup.find_all('div', class_="item-description"):
        try:
            title =  element.find('span',class_="item-description-title").text.strip()
            des = element.find("span", class_="item-description-abstract").text.strip()
            hyper = element.a.get('href')
        except:
            des = "No description listed."
        
        csv_writer.writerow([title,des,hyper])
    
    page +=1
```
