```python
import requests
from tkinter import *

root = Tk()
root.title("IN IT TO WIN IT")
root.geometry("1000x1000")

genius = requests.get("https://the-trivia-api.com/api/questions?categories=film_and_tv&limit=10&region=US&difficulty=easy")
json_file = genius.json()

question=[]

for i in range(0,len(json_file)):
     question.append(json_file[i]['question'])
        
class Quiz:
    def __init__(self):
        self.question_num=0
        self.gen_question()
        self.radio=StringVar()
        self.opts = self.gen_answers()
        self.buttons()
        self.dataSize = 10
        self.correct=0
        self.frame = Frame(root)
        
    def display_results(self):
        correct = f"Correct: {self.correct}"
        result=Label(root, text = correct)
        result.pack()
        #to show the number of questions that player got right.
                
    def selection(self,question_num):
        ans = self.radio.get()

        if ans == json_file[question_num]['correctAnswer']:
            return True
              
    def next_btn(self):
        if self.selection(self.question_num):
            self.correct+=1
            #if the chosen selection is the answer, it adds a point.
        self.question_num+=1
        #next question.
        if self.question_num==self.dataSize:
            self.display_results()
            root.destroy()
        else:
            self.frame.destroy()
            self.gen_question()
            self.gen_answers()
    
    def buttons(self):
        next_btn = Button(root, text = "NEXT",command=self.next_btn)
        next_btn.pack(side = BOTTOM)
        #creates the next button that shows the next question.
        
    def gen_question(self):
        self.frame = Frame(root)
        self.frame.pack(side = TOP,expand=False,fill="both")
        question_text = json_file[self.question_num]['question']
        question = Label(self.frame, text = '',font = ('arial 10'))
        question.config(text=question_text)
        question.pack(side=TOP)
        
        #prints the next question.
    
    def gen_answers(self):
        answers = []
        answers.append(json_file[self.question_num]['correctAnswer'])
        for i in range(0,len(json_file[self.question_num]['incorrectAnswers'])):
            answers.append(json_file[self.question_num]['incorrectAnswers'][i])
        r1 = Radiobutton(self.frame, text = '',variable = self.radio,value=answers[0],font = ('arial 7'))
        r1.config(text=answers[0])
        r1.pack(anchor = W)
        r2 = Radiobutton(self.frame, text = '',variable = self.radio,value=answers[1],font = ('arial 7'))
        r2.config(text=answers[1])
        r2.pack(anchor = W)
        r3 = Radiobutton(self.frame, text = '',variable = self.radio,value=answers[2],font = ('arial 7'))
        r3.config(text=answers[2])
        r3.pack(anchor = W)
        r4 = Radiobutton(self.frame, text = '',variable = self.radio,value=answers[3],font = ('arial 7'))
        r4.config(text=answers[3])
        r4.pack(anchor = W)
        return answers
        #the next set of radio buttons.

quiz=Quiz()
root.mainloop()

# with help from this website:
#https://www.geekedu.org/blogs/python-for-kids-python-quiz-game
```
